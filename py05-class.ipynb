### staticmethod
- 첫 번째 인자로 self, cls를 받지 않는다.
- 클래스나 인스턴스와 관계없는 일반적인 유틸리티 함수를 정의할때 사용한다.
- 독립적인 기능이지만 클래스 안에 넣는 것이 구조적으로 더 나을때 사용한다.

| 항목         | classmethod             | staticmethod |
| ---------- | ----------------------- | ------------ |
| 첫 번째 인자    | `cls` (클래스)             | 없음           |
| 클래스 변수 접근  | 가능                      | 불가능          |
| 인스턴스 변수 접근 | 불가능                     | 불가능          |
| 주 용도       | 클래스 상태를 사용하는 메서드        | 유틸리티 함수 정의   |
| 호출 방식      | 클래스.메서드() 또는 인스턴스.메서드() | 동일           |


````python
class B:
  x = 1


  def xx(self, x): # 인스턴스 메소드
    self.x = x


  @classmethod
  def yy(cls, x):  # 클래스 메소드
    cls.x = x


  @staticmethod
  def zz(a, b):  # 스테이틱 메소드
    print(a+b)


  @staticmethod  # 스테이틱 메소드
  def kk():
    print(x)
````

````python
B.yy(10)
````

````python
'''
[출력결과]
파이썬의정석의 저자는 김파이썬이고 150페이지로 구성됨
책이름 :  파이썬 입문
책저자 :  박북스
페이지 :  200 페이지
파이썬 입문의 저자는 박북스이고 200페이지로 구성됨
'''

class Book:
    def __init__(self, title=None, author=None, page=None):
        self.__title = title        # __ : private (직접 사용을 지양)
        self.__author = author
        self.__page = page

    def getTitle(self) :
        return self.__title

    def getAuthor(self) :
        return self.__author

    def getPage(self) :
        return self.__page

    def setTitle(self, title) :
        self.__title = title

    def setAuthor(self, author) :
        self.__author = author
    
    def setPage(self, page):
        self.__page = page

    def toString(self):
        return f"{self.__title}의 저자는 {self.__author}이고 {self.__page}페이지로 구성됨"
    
````

````python
class Dog:
    count = 0

    def __init__(self,name):
        self.name = name
        Dog.count = Dog.count + 1

    @classmethod        # 클래스메소드 데코레이터       #데코레이터가 없으면 cls는 self 용도로 사용이 됨 => count 호출 불가능
    def get_count(cls) :
        return cls.count

    @staticmethod
    def is_valid_name(name) :
        # name의 클래스가 문자열이고 길이는 0보다 크면 된다.
        return isinstance(name, str) and len(name) > 0
````

````python
from abc import ABC, abstractmethod


class Shape(ABC):


    def __init__(self, width, height):
        self.width = width
        self.height = height


    @abstractmethod
    def area(self):
      pass    




class Rect(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)


    def area(self):
      return self.width * self.height  


   
class Triangle(Shape):
   
    def __init__(self, width, height ):
        super().__init__(width, height)


    def area(self):
      return self.width * self.height * 0.5
````
