# 식별 연산자 (Identity Operators)
- 식별 연산자는 두 객체가 같은 객체인지를 확인한다.
- is : 두 객체가 같은 객체인지 확인
- is not 두 객체가 다른 객체인지 확인
- 결과 : True, False

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]
print(a is b)  # True (a와 b는 같은 객체)
print(a is c)  # False (a와 c는 같은 값을 가지지만 다른 객체)

print(a==b)
print(b==c)
```

### Generator Expression
 - 파이썬에는 tuple(immutable)에는 comprehension이 존재하지 않고 syntax 상으로 generator 가 생성한다.
 - 값을 하나씩 “생성”하는 반복자(iterator). 
 - 메모리를 덜 사용하며, **지연 평가(Lazy Evaluation)**가 특징이다.
 - generator를 호출하면 next 메소드를 통해 하나씩 호출하며 StopIteration 이 발생하면 종료된다.

 지연 평가 : 호출할 때마다 (필요할 때마다) 가져옴.

```python
squares = (x **2 for x in range(5) )    #값 생성 X. 모양만 만듦
print(squares)  #이건 제너레이터 객체 생성. (실행이 아님)
print(dir(squares))

#제너레이터는 결과가 바로 만들어지지 않음?
```

```python
#type( squares.__next__ )

for i in squares :
    print(i)
```

```python
squares = (x**2 for x in range(5) )
print(next(squares))
```

반복문에서의 yield 사용
  - for문 내부에서는 자동으로 next()를 호출하고, StopIteration이 발생하면 반복 종료합니다.

```python
def count_up_t(n) :
    i = 1
    while i <= n :
        yield i    #generator로 돌리겠단 뜻.
        i+=1
        #print(i)

data = count_up_t(3)
print(data)
```

```python
'''
만약 1,000만 개의 데이터 중 연산 비용이 아주 큰 작업(예: time.sleep이 포함되거나 복잡한 수학 연산)을 수행하다가, 
원하는 조건(예: 3의 배수)을 가진 첫 번째 결과만 찾고 싶을 때 두 방식의 차이
'''

import time


# 무겁고 오래 걸리는 연산을 시뮬레이션하는 함수
def heavy_processing(n):
    # 실제로는 아주 복잡한 연산이 일어난다고 가정합니다.
    return n * 2


# 1. 조기 평가 (Eager Evaluation) 방식
def eager_approach(data):
    # 조건과 상관없이 1,000만 개 데이터 전체에 대해 heavy_processing을 먼저 '전부' 실행함
    result = [heavy_processing(x) for x in data]    # [] :: 리스트!
   
    # 그 후, 필요한 조건을 찾음
    for val in result:
        if val > 0 and val % 3 == 0:
            return val


# 2. 지연 평가 (Lazy Evaluation) 방식
def lazy_approach(data):
    # 이 시점에는 아무런 연산도 하지 않음 (계산식만 준비)
    result_generator = (heavy_processing(x) for x in data)  # 괄호 :: 제너레이터 표현식
   
    # 필요한 순간(next)에만 heavy_processing을 하나씩 실행함
    for val in result_generator:
        if val > 0 and val % 3 == 0:
            return val


# 테스트 데이터 (0부터 9,999,999까지)
data_list = range(10000000)


# [비교 시작]
# 1) 조기 평가는 1,000만 번의 heavy_processing을 모두 수행하느라 엄청난 시간이 걸림
#eager_approach(data_list)
start_time = time.time()
print(f"즉시 평가 결과: {eager_approach(data_list)}")
print(f"소요 시간: {time.time() - start_time:.6f}초")

# 2) 지연 평가는 단 4번(x=0,1,2,3)만 연산하고 x=3일 때 6(3의 배수)을 반환하며 즉시 종료!
start_time = time.time()
print(f"지연 평가 결과: {lazy_approach(data_list)}")
print(f"소요 시간: {time.time() - start_time:.6f}초")
```

```python
'''
예제 1: 뉴스 기사 요약용 키워드 필터링기 (Comprehension 활용)
- 목표: 주어진 뉴스 본문 리스트에서 특정 키워드를 포함한 문장만 필터링
'''

news_articles = [
    "ChatGPT is transforming the AI world.",
    "Bitcoin price hits a new high.",
    "Python dominates machine learning.",
    "Weather forecast shows heavy rain.",
]


keywords = ['AI', 'Python', 'machine learning']

#filtered = [article for article in news_articles if 'AI' in article]

key = [
    k
    for k in news_articles
    if any(key in k 
           for key in keywords)
    ]
#any는 한개만 True여도 True.
#모두 False여야 False.

print(key)
```
